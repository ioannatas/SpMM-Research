```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


# Combining 'SpMM' with both csr and ie data
data = {
    'SpMM-32-csr':[d for d in data_csr_32_d if not 0],
    'SpMM-32-ie': [d for d in data_ie_32_d if not 0],
    'SpMM-42-csr': [d for d in data_csr_42_d if not 0],
    'SpMM-42-ie': [d for d in data_ie_42_d if not 0],
    'SpMM-64-csr': [d for d in data_csr_64_d if not 0],
    'SpMM-64-ie': [d for d in data_ie_64_d if not 0],
    'SpMM-128-csr': [d for d in data_csr_128_d if not 0],
    'SpMM-128-ie': [d for d in data_ie_128_d if not 0],
    'SpMM-256-csr': [d for d in data_csr_256_d if not 0],
    'SpMM-256-ie': [d for d in data_ie_256_d if not 0],
    'SpMM-350-csr': [d for d in data_csr_350_d if not 0],
    'SpMM-350-ie': [d for d in data_ie_350_d if not 0],
}


# Creating DataFrame with the combined data
df = pd.DataFrame(data)

# Plotting
plt.rcParams['figure.figsize'] = [20, 6]
plt.figure(figsize=(20, 6))

# Customizing colors
colors = {'csr': 'blue', 'ie': 'green'}
box_colors = [colors[col.split('-')[-1]] for col in df.columns]


# Create a box plot
boxplot = df.plot(kind='box', patch_artist=True)

# Customizing box colors
for patch, color in zip(boxplot.artists, box_colors):
    patch.set_facecolor(color)

# Customizing labels and title
plt.title('Double MKL CSR vs IE', fontsize=14)
plt.ylabel('GFLOPs', fontsize=12)
plt.xticks(fontsize=10, rotation=90)
plt.yticks(fontsize=10)

# Customizing legend
handles = [plt.Line2D([0], [0], color='blue', lw=4),
           plt.Line2D([0], [0], color='green', lw=4)]
labels = ['CSR', 'IE']
plt.legend(handles, labels, fontsize=10)

# Show the plot
plt.show()
```
